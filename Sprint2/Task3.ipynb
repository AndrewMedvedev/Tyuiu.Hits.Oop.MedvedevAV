```


using System;

    public class OrderLine
    {
        public int ProductId { get; set; }
        public string ProductName { get; set; }
        public decimal Price { get; set; }
        public DateTime DateAdded { get; set; }
        public string ProductCategory { get; set; }

        public virtual decimal CalculateTotal()
        {
        return Price;
        }

        public virtual void UpdatePrice(decimal newPrice)
        {
            Price = newPrice;
        }

        public virtual string GetProductDetails()
        {
            return $"Product ID: {ProductId}, Name: {ProductName}, Price: {Price:C}, Date Added: {DateAdded}, Category: {ProductCategory}";
        }

        public string GetSummary()
        {
            return $"Summary: {ProductName} - {Price:C}";
        }
    }



    public class StandardLine : OrderLine
    {
        public int Units { get; set; }

        public override decimal CalculateTotal()
        {
            return Price * Units;
        }

        public override string GetProductDetails()
        {
            return $"{base.GetProductDetails()}, Units: {Units}, Total: {CalculateTotal():C}";
        }

        public string GetUnits()
        {
            return $"Units in stock: {Units}";
        }
    }



    public class SpecialLine : OrderLine
    {
        public decimal Discount { get; set; }

        public override void UpdatePrice(decimal newPrice)
        {
            Price = newPrice * (1 - Discount / 100);
        }

        public bool IsDiscountActive(DateTime currentDate)
        {
            return (DateTime.Now - DateAdded).Days <= 30;
        }

        public override string GetProductDetails()
        {
            return $"{base.GetProductDetails()}, Discount: {Discount}%, Effective Price: {Price:C}, Discount Active: {IsDiscountActive(DateTime.Now)}";
        }
    }



    public class FreeLine : OrderLine
    {
        public decimal Prepayment { get; set; }

        public override decimal CalculateTotal()
        {
            return Price - Prepayment;
        }

        public string GetPrepaymentDetails()
        {
            return $"Prepayment: {Prepayment:C}";
        }

        public override string GetProductDetails()
        {
            return $"{base.GetProductDetails()}, Prepayment: {Prepayment:C}, Total after Prepayment: {CalculateTotal():C}";
        }
    }


    class Program
    {
        static void main(string[] args)
        {
            OrderLine standardLine = new StandardLine
            {
                ProductId = 1,
                ProductName = "Standard Product",
                Price = 10.00m,
                Units = 5,
                DateAdded = DateTime.Now,
                ProductCategory = "General"
            };

            OrderLine specialLine = new SpecialLine
            {
                ProductId = 2,
                ProductName = "Special Product",
                Price = 20.00m,
                Discount = 15, // 15%
                DateAdded = DateTime.Now,
                ProductCategory = "Promotion"
            };

            OrderLine freeLine = new FreeLine
            {
                ProductId = 3,
                ProductName = "Free Product",
                Price = 30.00m,
                Prepayment = 10.00m,
                DateAdded = DateTime.Now,
                ProductCategory = "Gift"
            };

            // Храним все строки заказа в массиве базового типа
            OrderLine[] orderLines = { standardLine, specialLine, freeLine };

            foreach (var line in orderLines)
            {
                Console.WriteLine(line.GetProductDetails());
                Console.WriteLine(line.GetSummary());
                Console.WriteLine();
            }
        }
    }

StandardLine stdLine = new StandardLine();
Console.WriteLine(stdLine.GetProductDetails());
Console.WriteLine($"Total: {stdLine.CalculateTotal():C}");

SpecialLine specialLine = new SpecialLine();
specialLine.UpdatePrice(30.00m); 
Console.WriteLine(specialLine.GetProductDetails());
Console.WriteLine($"Total: {specialLine.CalculateTotal():C}");

FreeLine freeLine = new FreeLine();
Console.WriteLine(freeLine.GetProductDetails());
Console.WriteLine($"Total: {freeLine.CalculateTotal():C}");


OrderLine[] orderLines = new OrderLine[] { stdLine, specialLine, freeLine };


decimal grandTotal = 0;
foreach (var orderLine in orderLines)
{
grandTotal += orderLine.CalculateTotal();
}
Console.WriteLine($"Grand Total: {grandTotal:C}");
```
